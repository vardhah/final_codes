```python
import pandas as pd 
import numpy as np
from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt
import tensorflow as tf

#csvfile = "DataVMC10.csv"  #VMC

csvfile = "Data_avf10.csv"  #AVF 100
csvfile="Data.csv"
#csvfile="Data_test.csv"
```

```python
data_test= pd.read_csv(csvfile)
data_test.head()
```

```python
profile = ProfileReport(data_test, minimal=True)
profile.to_file(output_file="output.html")
```

```python
failed = data_test[data_test['Stop_Distance'] <= 0]
passed = data_test[data_test['Stop_Distance'] > 0]

print('Size of crashes :',failed.shape[0])  
print('Size of success :',passed.shape[0]) 
print("Percentage of crash cases:{:.2f}".format(failed.shape[0]*100/(failed.shape[0]+passed.shape[0])),"%")
print("Percentage of not crashed cases:{:.2f}".format(passed.shape[0]*100/(failed.shape[0]+passed.shape[0])),"%")

failed.to_csv("crashed_gmm100.csv",index=False)
#failed.head()
```

```python
fig=plt.figure
plt.scatter(failed["Kick_Speed"],failed["Stop_Distance"],c = 'red',label='Line 2')
plt.scatter(passed["Kick_Speed"],passed["Stop_Distance"],c = 'grey',label='Line 1')
plt.grid(color='r', linestyle='-', linewidth=0.5)
plt.title("Failed vs pass cases")
plt.xlabel("Intial speed")
plt.ylabel("stopping distnace")
plt.show()
```

```python
fig=plt.figure
plt.scatter(passed["Kick_Speed"],passed["Stop_Distance"])
plt.grid(color='r', linestyle='-', linewidth=0.5)
plt.title("Success cases")
plt.xlabel("Intial speed")
plt.ylabel("distnace")
plt.title('Vanilla_testing_data-success')
plt.savefig('Vanilla_test1.png')
plt.show()
```

```python
data_test.loc[data_test['Stop_Distance'] <= 0, 'test'] = int(1)
data_test.loc[data_test['Stop_Distance'] > 0, 'test'] = int(0)
data_test.head()
```

```python
print('Total number of episodes:',data_test['Episode'].shape[0])
start=0
end=data_test['Episode'].shape[0]

fig, ax = plt.subplots()
episode=np.array(data_test['Episode'].values.tolist()[start:end])
test=np.array(data_test['test'].values.tolist()[start:end])
speed=np.array(data_test['Kick_Speed'].values.tolist()[start:end])

indices = test == 0
print('Number of Pass:',indices.sum())
print('Number of fail:',indices.shape[0]-indices.sum())

# normal values
ax.scatter(episode[~indices], speed[~indices], c = 'red',label='fail')
# bad values
ax.scatter(episode[indices], speed[indices], c = 'grey',label='Pass')
#ax.legend()
#ax.set_ylim([0,95])
plt.xlabel('episode')
plt.ylabel('speed (miles/hr)')
ax.grid(True) 
plt.title('AVF_500 failure search-episode')
plt.savefig('AVF_100fails.png')
plt.show()
```

```python
csvfile='Data_avf100_crashed.csv'
data_fail= pd.read_csv(csvfile)
total=data_fail['Episode'].shape[0]
fail_episode=data_fail['Episode'].values
print(fail_episode)
failure_episode=[];
for i in range(total):
    if i==0: 
        failure_episode.append(fail_episode[i])
    else: 
        failure_episode.append(fail_episode[i]-fail_episode[i-1]) 
    
print('The fail episode :',failure_episode)
print('length of failures: ',len(failure_episode))
running_avg=[];
store=0
for i in range(total):
        store=(store+failure_episode[i])
        running_avg.append(round(store/(i+1),1))
print('running avg :',running_avg)
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(12, 8), dpi=300, facecolor='w', edgecolor='k')
```

```python
fig, ax = plt.subplots()
ax.plot(failure_episode,c = 'green',label='failures',marker='o',linewidth=1, markersize=6) 
ax.plot(running_avg,c = 'orange',label='running average')
plt.xlabel('failures')
plt.ylabel('simulation episodes')
ax.legend()
ax.grid(True) 
plt.title('AVF_100 failure search-number of episodes')
plt.savefig('AVF_100fails.png')
plt.show()
```

```python
csvfile='Data_vmc100_crashed.csv'
data_fail= pd.read_csv(csvfile)
total=data_fail['Episode'].shape[0]
fail_episode=data_fail['Episode'].values
print(fail_episode)
failure_episode_vmc=[];
for i in range(total):
    if i==0: 
        failure_episode_vmc.append(fail_episode[i])
    else: 
        failure_episode_vmc.append(fail_episode[i]-fail_episode[i-1]) 
    
print('The fail episode :',failure_episode_vmc)
print('length of failures: ',len(failure_episode_vmc))
running_avg_vmc=[];
store=0
for i in range(total):
        store=(store+failure_episode_vmc[i])
        running_avg_vmc.append(round(store/(i+1),1))
print('running avg :',running_avg_vmc)
```
